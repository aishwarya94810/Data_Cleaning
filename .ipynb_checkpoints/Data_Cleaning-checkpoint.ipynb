CLEANING THE data.csv FILE USING VARIOUS PYTHON LIBRARIES. 

Importing libraries needed for cleaning the data.

```python
import pandas as pd
import numpy as np
import re as regex
from datetime import datetime as dt
```

Imporing the csv file using the pandas function

```python
edata = pd.read_csv('data.csv')
edata
```

Date Column-Changing the row values to NaT that are not in date format(redundant values).
Replacing all the empty row values of Customer_ID with NaN.
Removing the alphabets from the InvoiceNo column and considering only digits. Example: A571035 -> 571035
Sort the Planet column alphabetically A-Z

```python
edata.drop(edata.columns[0], axis=1)
edata['Date'] = pd.to_datetime(edata['Date'], errors='coerce')
edata['Customer_ID'] = edata['Customer_ID'].replace(r'\s+', np.nan, regex=True)
edata['InvoiceNo'] = edata['InvoiceNo'].str.extract('(\d+)', expand=False)
edata.sort_values('Planet',inplace=True, ascending=True)

edata1 = edata
edata1
```

Some rows contain redundant values that do not belong to any of the 5 columns(Which resemble the Purchase column). 
Removed those rows along with any redunandt rows.

```python
edata2 = edata1[pd.to_numeric(edata1.Customer_ID,errors='coerce').notnull()]
edata2
```

```python
data1 = pd.DataFrame(edata2, columns = ['InvoiceNo','Customer_ID','Date','Planet'])
data1.to_csv(r'data1.csv')
```

Further I'm forming a second table called purchase which contains InvoiceNo and Purchase.
This purchase table is further to be cleaned and linked to the data1 table on the InvoiceNo.
InvoiceNo is the primary key in the data1 and Foreign key in the purchase table.

```python
edata3 = pd.DataFrame(edata2, columns = ['InvoiceNo','Purchase'])
edata3
```

Cleaning the Purchase table and splitting the purchase rows which are list of dictionaries into columns of values.

```python
tmp_list = list()
for index, row in edata3.iterrows():
    for list_item in eval(row['Purchase']):
        tmp_list.append(dict(list(list_item.items())+[('InvoiceNo', row['InvoiceNo'])]))
```

```python
df1=pd.DataFrame(tmp_list)
df1
```

Saving the above dataframe into purchase.csv

```python
df1.to_csv(r'purchase.csv')
```
